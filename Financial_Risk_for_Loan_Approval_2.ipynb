```python
numerical_columns = ['Age', 'AnnualIncome', 'CreditScore', 'Experience', 'LoanAmount', 'LoanDuration', 'NumberOfDependents', 
                     'MonthlyDebtPayments', 'CreditCardUtilizationRate','LengthOfCreditHistory', 'SavingsAccountBalance', 
                     'CheckingAccountBalance', 'JobTenure', 'NetWorth', 'NumberOfOpenCreditLines', 'NumberOfCreditInquiries', 
                     'DebtToIncomeRatio', 'TotalAssets', 'TotalLiabilities', 'MonthlyIncome', 'BaseInterestRate', 'InterestRate', 
                     'MonthlyLoanPayment', 'TotalDebtToIncomeRatio', 'RiskScore', 'PaymentHistory', 'BankruptcyHistory', 
                     'UtilityBillsPaymentHistory', 'PreviousLoanDefaults', 'LoanApproved']

categorical_columns = ['ApplicationDate', 'EmploymentStatus', 'EducationLevel', 'MaritalStatus', 'HomeOwnershipStatus', 'LoanPurpose']
```

```python
binary_columns = []
discrete_columns = []
continuous_columns = []

for col in numerical_columns:
    unique_values = data[col].dropna().unique()

    if set(unique_values).issubset({0, 1}):
        binary_columns.append(col)

    elif all(float(x).is_integer() for x in unique_values):
        discrete_columns.append(col)

    else:
        continuous_columns.append(col)

print("Kolumny binarne:", binary_columns)
print("Kolumny dyskretne:", discrete_columns)
print("Kolumny ciągłe:", continuous_columns)
```

```python
binary_columns = ['BankruptcyHistory', 'PreviousLoanDefaults', 'LoanApproved']
discrete_columns = ['Age', 'AnnualIncome', 'CreditScore', 'Experience', 'LoanAmount', 'LoanDuration', 
                    'NumberOfDependents', 'MonthlyDebtPayments', 'LengthOfCreditHistory', 
                    'SavingsAccountBalance', 'CheckingAccountBalance', 'JobTenure', 'NetWorth', 
                    'NumberOfOpenCreditLines', 'NumberOfCreditInquiries', 'TotalAssets', 
                    'TotalLiabilities', 'PaymentHistory']
continuous_columns = ['CreditCardUtilizationRate', 'DebtToIncomeRatio', 'MonthlyIncome', 
                      'BaseInterestRate', 'InterestRate', 'MonthlyLoanPayment', 
                      'TotalDebtToIncomeRatio', 'RiskScore', 'UtilityBillsPaymentHistory']
categorical_columns = ['EmploymentStatus', 'EducationLevel', 'MaritalStatus', 
                       'HomeOwnershipStatus', 'LoanPurpose', 'ApplicationDate']
```

# Dataset 1: Wypełnienie medianą/modą

```python
data1 = data.copy()

# Wypełnianie braków w kolumnach binarnych (modą)
for col in binary_columns:
    mode_value = data1[col].mode()[0]
    data1[col].fillna(mode_value, inplace=True)

# Wypełnianie braków w kolumnach dyskretnych (medianą) i zaokrąglanie do liczb całkowitych
for col in discrete_columns:
    median_value = data1[col].median()
    data1[col].fillna(median_value, inplace=True)
    data1[col] = data1[col].round().astype(int)

# Wypełnianie braków w kolumnach ciągłych (medianą)
for col in continuous_columns:
    median_value = data1[col].median()
    data1[col].fillna(median_value, inplace=True)

# Wypełnianie braków w kolumnach kategorycznych (modą) i ustawianie typu jako kategoria
for col in categorical_columns:
    mode_value = data1[col].mode()[0]
    data1[col].fillna(mode_value, inplace=True)
    data1[col] = data1[col].astype('category')

data1.to_csv('dataset_filled_median.csv', index=False)
```

# Dataset 2: Wypełnienie średnią/modą

```python
data2 = data.copy()

# Wypełnianie braków w kolumnach binarnych (modą)
for col in binary_columns:
    mode_value = data2[col].mode()[0]
    data2[col].fillna(mode_value, inplace=True)

# Wypełnianie braków w kolumnach dyskretnych (średnią) i zaokrąglanie do liczb całkowitych
for col in discrete_columns:
    mean_value = data2[col].mean()
    data2[col].fillna(mean_value, inplace=True)
    data2[col] = data2[col].round().astype(int)

# Wypełnianie braków w kolumnach ciągłych (średnią)
for col in continuous_columns:
    mean_value = data2[col].mean()
    data2[col].fillna(mean_value, inplace=True)

# Wypełnianie braków w kolumnach kategorycznych (modą) i ustawianie typu jako kategoria
for col in categorical_columns:
    mode_value = data2[col].mode()[0]
    data2[col].fillna(mode_value, inplace=True)
    data2[col] = data2[col].astype('category')

data2.to_csv('dataset_filled_mean.csv', index=False)
```

# Dataset 3: Wypełnianie braków za pomocą regresji

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

selected_features = ['CreditScore', 'NetWorth', 'MonthlyIncome', 'BaseInterestRate', 'TotalDebtToIncomeRatio', 'LengthOfCreditHistory']
target_columns = ['AnnualIncome', 'LoanAmount', 'LoanDuration', 'MonthlyLoanPayment']

clean_data = data.dropna(subset=selected_features)
print(f"Liczba rekordów po oczyszczeniu: {len(clean_data)}")

X = clean_data[selected_features]
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

models = {}

for target in target_columns:
    if data[target].isnull().sum() > 0:
        y = clean_data[target].dropna()
        X_target = clean_data.loc[y.index, selected_features]
        X_target_poly = poly.transform(X_target)

        X_train, X_test, y_train, y_test = train_test_split(X_target_poly, y, test_size=0.2, random_state=42)

        model = LinearRegression()
        model.fit(X_train, y_train)
        models[target] = model

        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        r2_train = r2_score(y_train, y_train_pred)
        r2_test = r2_score(y_test, y_test_pred)
        mse_train = mean_squared_error(y_train, y_train_pred)
        mse_test = mean_squared_error(y_test, y_test_pred)

        print(f"Model dla {target} wytrenowany.")
        print(f"R² (trening): {r2_train:.2f}, R² (test): {r2_test:.2f}")
        print(f"MSE (trening): {mse_train:.2f}, MSE (test): {mse_test:.2f}")
```

```python
data3 = data.copy()

# 1. Wypełnianie braków za pomocą wytrenowanych modeli
for target in target_columns:
    if target in models:
        model = models[target]
        missing_mask = data3[target].isnull()
        if missing_mask.sum() > 0:
            print(f"Uzupełnianie braków w kolumnie: {target}")
            X_missing = data3.loc[missing_mask, selected_features].fillna(0)
            X_missing_poly = poly.transform(X_missing)
            predicted_values = model.predict(X_missing_poly)

            # Zaokrąglanie wartości dla zmiennych dyskretnych
            if target in ['LoanDuration']:
                predicted_values = np.round(predicted_values)

            # Uzupełnianie braków
            data3.loc[missing_mask, target] = predicted_values

# 2. Wypełnianie braków w kolumnach binarnych (moda)
for column in binary_columns:
    if data3[column].isnull().sum() > 0:
        mode_value = data3[column].mode()[0]
        data3[column].fillna(mode_value, inplace=True)

# 3. Wypełnianie braków w kolumnach kategorycznych (moda)
for column in categorical_columns:
    if data3[column].isnull().sum() > 0:
        mode_value = data3[column].mode()[0]
        data3[column].fillna(mode_value, inplace=True)

# 4. Wypełnianie braków w kolumnach numerycznych (połowa medianą, połowa średnią)
for column in discrete_columns + continuous_columns:
    if data3[column].isnull().sum() > 0:
        median_value = data3[column].median()
        mean_value = data3[column].mean()
        missing_indices = data3[column].isnull()
        half_indices = missing_indices[missing_indices].index[:len(missing_indices) // 2]
        data3.loc[half_indices, column] = median_value
        data3[column].fillna(mean_value, inplace=True)

# 5. Konwersja typów dla kolumn dyskretnych i binarnych
for column in discrete_columns:
    data3[column] = data3[column].astype(int)

for column in binary_columns:
    data3[column] = data3[column].astype(float)

data3.to_csv('dataset_filled_combined_final.csv', index=False)
```

# Skalowanie min-max

```python
#data = pd.read_csv('dataset_filled_mean.csv')
#data = pd.read_csv('dataset_filled_median.csv')
data = pd.read_csv('dataset_filled_combined_final.csv')

dataset4 = data.copy()

dataset4.drop(columns=['ApplicationDate'], inplace=True)

one_hot_columns = ['HomeOwnershipStatus', 'MaritalStatus', 'LoanPurpose']
label_encode_columns = ['EmploymentStatus', 'EducationLevel']

dataset4 = pd.get_dummies(dataset4, columns=one_hot_columns)
one_hot_encoded_columns = [col for col in dataset4.columns if any(prefix in col for prefix in one_hot_columns)]
dataset4[one_hot_encoded_columns] = dataset4[one_hot_encoded_columns].astype(int)

label_encoder = LabelEncoder()
for column in label_encode_columns:
    dataset4[column] = label_encoder.fit_transform(dataset4[column])

scaler = MinMaxScaler()

numerical_columns = [col for col in discrete_columns + continuous_columns if col not in binary_columns]
dataset4[numerical_columns] = scaler.fit_transform(dataset4[numerical_columns])

#dataset4.to_csv('dataset_filled_mean_min_max.csv', index=False)
#dataset4.to_csv('dataset_filled_median_min_max.csv', index=False)
dataset4.to_csv('dataset_filled_combined_min_max.csv', index=False)
print("Przetwarzanie zakończone i zapisane do pliku csv.")
```
